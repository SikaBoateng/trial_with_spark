```python
# Read the orders table
orders = pd.read_csv('update_mk2_orders.csv')

# Calculate aggregate metrics
aggregate_metrics = {
    'Total Transaction Amount per User': orders.groupby('Customer ID')['Order Total'].sum(),
    'Average Order Value': orders['Order Total'].mean(),
    'Total Sales Revenue': orders['Order Total'].sum(),
    'Total Number of Orders': orders['Order ID'].nunique(),
    'Total Number of Customers': orders['Customer ID'].nunique(),
    'Total Quantity Sold per Category': orders.groupby('Category Name')['Quantity'].sum(),
    'Total Revenue per Category': orders.groupby('Category Name')['Order Total'].sum(),
    'Average Transaction Time': (pd.to_datetime(orders['Delivery Time']) - pd.to_datetime(orders['Order Time'])).mean(),
    'Total Revenue per Merchant': orders.groupby('Merchant ID')['Order Total'].sum(),
    'Total Revenue per Payment Method': orders.groupby('Payment Method')['Order Total'].sum()
}

# Display the aggregate metrics
for metric, result in aggregate_metrics.items():
    print(metric)
    print(result)
    print()
```
